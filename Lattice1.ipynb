```python
import numpy as np
from numpy import linalg as la
import matplotlib.pyplot as plt
import networkx as nx
```

```python
g = nx.Graph()
g.add_node(0)
```

```python
g.add_node(1); g.add_node(2); g.add_node(3)
```

```python
g.add_edge(0,1); g.add_edge(1,2); g.add_edge(2,3); g.add_edge(3,0)
g.edges()
```

```python
nx.draw(g)
```

```python
c = nx.cycle_graph(4)
nx.draw(c)
```

```python
a = nx.adjacency_matrix(c)
print(a.todense())
type(a.todense())
```

```python
w, v = la.eig(np.array(a.todense()))
print(w)
```

```python
g1 = nx.grid_graph([4,4])
nx.draw(g1)
```

```python
w1, v1 = la.eig(np.array(nx.adjacency_matrix(g1).todense()))
print(w1)
```

```python
line_g1 = nx.line_graph(g1)
nx.draw(line_g1)
```

```python
w1_l, v1_l = la.eig(np.array(nx.adjacency_matrix(line_g1).todense()))
print(w1_l)
```

```python
plt.plot(range(1, len(w1_l) + 1), np.sort(np.real(w1_l)), 'ro')
plt.xlabel('Eigenvalue index')
plt.ylabel('Eigenvalue')
plt.title('4x4 Square Layout Spectrum')
#plt.rcParams["figure.figsize"] = [16,9]
plt.show()
```

```python
hex = nx.hexagonal_lattice_graph(3,3)
nx.draw(hex)
```

```python
W = nx.Graph()
W.add_edge(0,1, weight = 0.5); W.add_edge(1,2, weight = 1); W.add_edge(2,3, weight=1); W.add_edge(3,0, weight=1)
nx.draw(W)
```

```python
a_W = nx.adjacency_matrix(W)
print(a_W.todense())
```

```python
w_aW, v_aW = la.eig(np.array(a_W.todense()))
print(w_aW)
```

```python
line_hex = nx.line_graph(hex)
w_hex, v_hex = la.eig(np.array(nx.adjacency_matrix(line_hex).todense()))
```

```python
plt.plot(range(1, len(w_hex) + 1), np.sort(np.real(w_hex)), 'ro')
plt.xlabel('Eigenvalue index')
plt.ylabel('Eigenvalue')
plt.title('3x3 Hexagonal Layout Spectrum')
plt.rcParams.update({'font.size': 29})
plt.show()
```
